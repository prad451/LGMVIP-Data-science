**Image to pencil sketch with python**

**Importing liabraries**

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

**Loading and Plotting Original Image**

```python
img=cv2.imread('/content/puppy.jpg.png')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.title("Original Image")
plt.imshow(img)
```

**Converting Original Image to GrayScale**

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(img_gray,cmap="gray")
plt.title("GrayScale Image")
plt.show()
```

**Inverting the Image**

```python
img_invert=cv2.bitwise_not(img_gray)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(img_invert,cmap="gray")
plt.title("Inverted Image")
plt.show()
```

**Smoothing the Image**

```python
img_smoothing=cv2.GaussianBlur(img_invert, (21,21), sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(img_smoothing,cmap="gray")
plt.title("Smoothen Image")
plt.show()
```

**Converting Image to Pencil Sketch**

```python
img_sketch=cv2.divide(img_gray, 255-img_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(img_sketch,cmap="gray")
plt.title("Final Sketch Image")
plt.show()
```

**Displaying all the Images**

```python
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.subplot(1,5,5)
plt.imshow(img_sketch,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
